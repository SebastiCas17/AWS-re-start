```python
print("Welcome to Guess the Number!")

number = random.randint(1, 10)
isGuessRight = False

while not isGuessRight:
    guess = int(input("Guess a number between 1 and 10: "))

    if guess == number:
        print("Correct! You win!")
        isGuessRight = True
    else:
        print("Wrong number, try again.")

print("Count to 10")

for x in range(11):
    print(x)
```
